```python
def laplacecomp(w, X, t):
    '''
        Return two np array for g and log(g)
    '''
    # Computes g and log g for the laplace model introduced in the lecture.
    ss = 10 #% Prior variance
    # Evaluate log prior
    logg = -(1/(2 * ss)) * w.transpose().dot(w)
    # Compute P
    P = np.array(1. / ( 1 + np.exp(-X.dot(w))))
    #print(P)
    logl = np.sum((t * np.log(P)) + ((1 - t) * np.log(1-P)))
    
    logg = logg + logl
    g = np.exp(logg)
    return (g[0][0], logg[0][0])
```

```python
def acceptance_ratio(w_s, w_s1, X, t):
    #print(w_s)
    r = laplacecomp(w_s, X, t)[0] / laplacecomp(w_s1, X, t)[0]
    #print(laplacecomp(w_s, X, t)[0])
    return r
```

```python
def MH(X, t, total_trials, step):
    s = 0
    w = []
    w.append(np.ones([X.shape[1], 1]))
    #print(w[-1])
    while s < total_trials:
        if (s+1) % (total_trials/100) == 0:
            if (s+1) % (total_trials/10) == 0:
                print('|', end='')
            else:
                print('.', end='')
        s += 1
        #Generate ~w_s
        new_w_s = np.random.multivariate_normal(mean=w[-1][:,0], cov=step)
        new_w_s = np.array([new_w_s]).T
        #new_w_s = np.random.rand(2, 1) * 0.5 + w[-1]
        #print(new_w_s)
        #new_w_s = np.array([new_w_s]).T
        #print(new_w_s)
        #Compute acceptance ratio
        r = (acceptance_ratio(new_w_s, w[-1], X, t))
        
        #print(r)
        #Test r
        if r >= 1:
            w.append(new_w_s)
            continue
        else:
            u = np.random.uniform(low=0, high=1)
            #print(u)
            if u <= r:
                w.append(new_w_s)
                continue
            else:
                w.append(w[-1])
    print()
    return w
```

```python
step = np.array([[1, 0], [0, 1]])
num_trials = 100000
samples = MH(X, t, num_trials, step)
#pprint(samples)
```
